```python
import requests
import pandas as pd

def fetch_tabnet_data(disease_code, city_code):
    base_url = 'http://tabnet.datasus.gov.br/cgi/tabcgi.exe?sih/cnv/qiSP.def'
    
    # Parâmetros da requisição
    params = {
        'qtdade': 'mínima',
        'cbo02': '00',
        'cbo01': '00',
        'cbo07': '00',
        'cbo06': '71',
        'cbo05': '00',
        'cbo04': '00',
        'cbo03': '00',
        'cbo12': '02',
        'cbo11': '00',
        'cbo10': '00',
        'cbo09': '00',
        'cbo08': '00',
        'cc': "J44.9",  # Código da doença (DPOC - J44.9)
        'lc': "3550308",     # Código da cidade de São Paulo (3550308)
        'line': 'on',
        'expurgos': 'on',
        'sigla': 'SP',
        'qtdn': 'n',
        'frames': 'a',
        'f': 'd70',
    }
    
    try:
        # Faz a requisição GET
        response = requests.get(base_url, params=params, timeout=60)
        
        # Verifica se a requisição foi bem-sucedida
        if response.status_code == 200:
            # Lê o conteúdo da resposta em um DataFrame pandas
            df = pd.read_csv(response.url, sep=';', encoding='ISO-8859-1', skiprows=3, skipfooter=12, engine='python')
            
            return df
        
        else:
            print(f'Erro ao obter dados: Código {response.status_code}')
            return None
    
    except requests.exceptions.RequestException as e:
        print(f'Erro de requisição: {e}')
        return None

def main():
    disease_code = 'J44.9'  # Código para DPOC
    city_code = '3550308'   # Código IBGE da cidade de São Paulo
    
    # Chamada para obter os dados
    data = fetch_tabnet_data(disease_code, city_code)
    
    # Se os dados foram obtidos com sucesso, faça algo com eles
    if data is not None:
        print('Dados obtidos com sucesso:')
        print(data.head())  # Mostra as primeiras linhas do DataFrame
        # Aqui você pode processar os dados conforme necessário
```

```python
import requests
from datetime import datetime
import json
import pandas as pd


def get_api_data():
    #end point API
    API_URL = "https://api.waqi.info"
    #token de acesso
    token = "77d05ca5845b52922ee22f41d4386752adb7e9f5"
    #chave de cidade de medicao
    keyword = "Osasco, São Paulo, Brazil"
    #cidade medicao
    city = "osasco"

    #requisicao get para dados de medicao horaria
    jsson = feed(API_URL, token, city)

    # json_object = json.loads(jsson)

    print(json.dumps(jsson, indent=2))

    # print(pd.json_normalize(jsson['pm10']))

    return jsson

def search(API_URL, token, keyword):   
    search_url = f"{API_URL}/v2/search/?token={token}&keyword={keyword}"
    response = requests.get(search_url)
    result = response.json()
    
    return (result)

def feed(API_URL, token, city):
    """
    retorna medicao horaria e previsao de AQI
    """
    search_url = f"{API_URL}/feed/{city}?token={token}"
    response = requests.get(search_url)
    result = response.json()
    
    return (result)
```
